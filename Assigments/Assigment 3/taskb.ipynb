```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib widget

df = pd.read_csv('/home/ubuntu/comp_astro_24/Assigments/Assigment 3/table_WASP-39-b-Carter-et-al.-2024.csv')

df = df[['CENTRALWAVELNG', 'BANDWIDTH', 'PL_TRANDEP', 'PL_TRANDEPERR1']]

df.to_csv('/home/ubuntu/comp_astro_24/Assigments/Assigment 3/Wasp39b_spectrum_assignment3_taskB.csv')

# Estrai i dati per il plot
wavelength = df['CENTRALWAVELNG']  # Lunghezza d'onda in micron
wavelength_error = df['BANDWIDTH'] # Errore su lunghezza d'onda
transit_depth = df['PL_TRANDEP']   # Profondit di transito (Rp/Rs)^2
error = df['PL_TRANDEPERR1']       # Errore sulla profondit di transito

# Crea una figura pi larga
plt.figure(figsize=(12, 2))  # Aumenta il valore della larghezza (primo parametro)

# Crea il grafico dello spettro
plt.errorbar(wavelength, transit_depth, xerr=wavelength_error, yerr=error, fmt='o', markersize=4, capsize=2, label='Dati osservati')

# Configura il grafico
plt.xlabel('Lunghezza d\'onda [$\\mu$m]')
plt.ylabel('Profondit di transito $(R_p / R_s)^2$')
plt.title('Spettro di trasmissione di WASP-39b')
plt.legend()
plt.grid(True)

# Salva il grafico in un file PNG
plt.savefig('/home/ubuntu/comp_astro_24/Assigments/Assigment 3/Wasp39b_spectrum_assignment3.png', dpi=300)

# Mostra il grafico
plt.show()
```
